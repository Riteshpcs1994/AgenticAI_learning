```python
from sentence_transformers import SentenceTransformer
from sentence_transformers.util import cos_sim
import numpy as np

# Load model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Query and passages
query = "How many people live in Berlin?"
passages = [
    "Berlin had a population of 3,520,031 registered inhabitants in an area of 891.82 square kilometers.",
    "Berlin is well known for its museums.",
    "In 2014, the city state Berlin had 37,368 live births (+6.6%), a record number since 1991.",
    "The urban area of Berlin comprised about 4.1 million people in 2014, making it the seventh most populous urban area in the European Union.",
    "The city of Paris had a population of 2,165,423 people within its administrative city limits as of January 1, 2019",
    "An estimated 300,000-420,000 Muslims reside in Berlin, making up about 8-11 percent of the population.",
    "Berlin is subdivided into 12 boroughs or districts (Bezirke).",
    "In 2015, the total labour force in Berlin was 1.85 million.",
    "In 2013 around 600,000 Berliners were registered in one of the more than 2,300 sport and fitness clubs.",
    "Berlin has a yearly total of about 135 million day visitors, which puts it in third place among the most-visited city destinations in the European Union.",
]

# Ground truth relevance labels (0 = not relevant, 1 = somewhat relevant, 2 = highly relevant, 3 = perfectly relevant)
# You need to define these based on your use case
relevance_labels = [3, 0, 1, 2, 0, 2, 0, 1, 0, 1]  # Example labels

# Encode query and passages
query_embedding = model.encode(query, convert_to_tensor=True)
passage_embeddings = model.encode(passages, convert_to_tensor=True)

# Compute similarity scores
similarity_scores = cos_sim(query_embedding, passage_embeddings)[0]

# Create ranking
ranks = [
    {"corpus_id": idx, "score": score.item()} 
    for idx, score in enumerate(similarity_scores)
]
ranks = sorted(ranks, key=lambda x: x["score"], reverse=True)

# Print ranked results
print("Query:", query)
print("\nRanked Results:")
for i, rank in enumerate(ranks):
    print(f"Rank {i+1}: {rank['score']:.2f}\t[Relevance: {relevance_labels[rank['corpus_id']]}]\t{passages[rank['corpus_id']][:80]}...")


# ============== MRR (Mean Reciprocal Rank) ==============
def calculate_mrr(ranked_results, relevance_labels, relevance_threshold=1):
    """
    Calculate MRR - finds the rank of the first relevant document
    
    Args:
        ranked_results: List of ranked results with 'corpus_id'
        relevance_labels: Ground truth relevance scores
        relevance_threshold: Minimum relevance score to consider as relevant
    """
    for rank_position, result in enumerate(ranked_results, start=1):
        corpus_id = result['corpus_id']
        if relevance_labels[corpus_id] >= relevance_threshold:
            return 1.0 / rank_position
    return 0.0

mrr = calculate_mrr(ranks, relevance_labels, relevance_threshold=2)
print(f"\n{'='*60}")
print(f"MRR (Mean Reciprocal Rank): {mrr:.4f}")
print(f"First highly relevant doc found at rank: {int(1/mrr) if mrr > 0 else 'N/A'}")


# ============== nDCG (Normalized Discounted Cumulative Gain) ==============
def calculate_dcg(relevance_scores, k=None):
    """
    Calculate DCG (Discounted Cumulative Gain)
    
    Args:
        relevance_scores: List of relevance scores in ranked order
        k: Consider only top-k results (None = all results)
    """
    if k is not None:
        relevance_scores = relevance_scores[:k]
    
    dcg = 0.0
    for i, rel in enumerate(relevance_scores, start=1):
        dcg += rel / np.log2(i + 1)
    return dcg

def calculate_ndcg(ranked_results, relevance_labels, k=None):
    """
    Calculate nDCG (Normalized DCG)
    
    Args:
        ranked_results: List of ranked results with 'corpus_id'
        relevance_labels: Ground truth relevance scores
        k: Consider only top-k results (None = all results)
    """
    # Get relevance scores in ranked order
    ranked_relevance = [relevance_labels[result['corpus_id']] for result in ranked_results]
    
    # Calculate DCG for the ranking
    dcg = calculate_dcg(ranked_relevance, k)
    
    # Calculate ideal DCG (best possible ranking)
    ideal_relevance = sorted(relevance_labels, reverse=True)
    idcg = calculate_dcg(ideal_relevance, k)
    
    # Avoid division by zero
    if idcg == 0:
        return 0.0
    
    return dcg / idcg

# Calculate nDCG at different cutoffs
ndcg_full = calculate_ndcg(ranks, relevance_labels)
ndcg_at_3 = calculate_ndcg(ranks, relevance_labels, k=3)
ndcg_at_5 = calculate_ndcg(ranks, relevance_labels, k=5)
ndcg_at_10 = calculate_ndcg(ranks, relevance_labels, k=10)

print(f"\nnDCG (full): {ndcg_full:.4f}")
print(f"nDCG@3: {ndcg_at_3:.4f}")
print(f"nDCG@5: {ndcg_at_5:.4f}")
print(f"nDCG@10: {ndcg_at_10:.4f}")


# ============== Additional Metrics ==============
def calculate_precision_at_k(ranked_results, relevance_labels, k, relevance_threshold=1):
    """Calculate Precision@k"""
    top_k = ranked_results[:k]
    relevant_count = sum(1 for result in top_k if relevance_labels[result['corpus_id']] >= relevance_threshold)
    return relevant_count / k

def calculate_recall_at_k(ranked_results, relevance_labels, k, relevance_threshold=1):
    """Calculate Recall@k"""
    top_k = ranked_results[:k]
    relevant_in_top_k = sum(1 for result in top_k if relevance_labels[result['corpus_id']] >= relevance_threshold)
    total_relevant = sum(1 for label in relevance_labels if label >= relevance_threshold)
    return relevant_in_top_k / total_relevant if total_relevant > 0 else 0.0

print(f"\nPrecision@3: {calculate_precision_at_k(ranks, relevance_labels, 3, relevance_threshold=2):.4f}")
print(f"Precision@5: {calculate_precision_at_k(ranks, relevance_labels, 5, relevance_threshold=2):.4f}")
print(f"Recall@3: {calculate_recall_at_k(ranks, relevance_labels, 3, relevance_threshold=2):.4f}")
print(f"Recall@5: {calculate_recall_at_k(ranks, relevance_labels, 5, relevance_threshold=2):.4f}")
```
